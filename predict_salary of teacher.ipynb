The aim of this project is to predict the salary of a teache based on the service he/she provide
for this we are using linear regression as we are predicting continuous value outputs 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('http://rcs.bu.edu/examples/python/data_analysis/Salaries.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
plt.scatter(df.service, df.salary)
plt.xlabel('service')
plt.ylabel('salary')
plt.show()
```

```python
plt.figure(figsize=(14,10))
plt.tight_layout()
sns.distplot(df['salary'].values)
```

Linear regression performs the task to predict a dependent variable value (y) based on a given independent variable (x). So, this regression technique finds out a linear relationship between x (input) and y(output).
The equation of the line is : Y= mx + b
where b = intercept , m= slope line

```python
df.columns
```

```python
X= df.service.values.reshape(-1,1)
y = df.salary.values.reshape(-1,1)
```

```python
X
```

```python
y
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = .3,random_state =0)
```

```python
regressor = LinearRegression()
```

```python
regressor.fit(X_train,y_train)
```

###### The intercept (often labeled the constant) is the expected mean value of Y when all X=0.

Start with a regression equation with one predictor, X.

If X sometimes equals 0, the intercept is simply the expected mean value of Y at that value.

```python
regressor.intercept_
```

```python
regressor.coef_
```

```python
y_pred = regressor.predict(X_test)
```

```python
df1 = pd.DataFrame({'Actual':y_test.flatten(), 'Predicted':y_pred.flatten()})
df1
```

```python
dfp = df1.head(20)
dfp.plot(kind='bar',figsize =(15,11))
plt.grid(which='major',linestyle =':', linewidth ='.7',color = 'green')
plt.grid(which = 'major', linestyle = ":", linewidth='.7',color = 'black')
plt.show()
```

```python
#Let's plot our straight line with the test data 
plt.scatter(X_test,y_test)
plt.plot(X_test,y_pred,color='red', linewidth = 3)
plt.show()
```

### The mean absolute error 


```python
from sklearn import metrics
import numpy as np
```

```python
print("The mean absolute error is :", metrics.mean_absolute_error(y_test,y_pred))
```

### The mean squared error 
formuma: 

```python
print("the mean squared error is :", metrics.mean_squared_error(y_test,y_pred))
```

###  The root mean square
formula:

```python
print("the root mean squared error is :", np.sqrt(metrics.mean_squared_error(y_test,y_pred)))
```

You can see that the value of root mean squared error is 17076.969831217742, which is more than the mean value of the percentages of all. This means that our algorithm was not very accurate but can still make reasonably good predictions.
